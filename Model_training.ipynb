```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, TextDataset, DataCollatorForLanguageModeling, Trainer, TrainingArguments
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
```

```python
train_dataset = TextDataset(
    tokenizer=tokenizer,
    file_path='/content/drive/MyDrive/nlp_final_project_1.txt',  # Replace with the path to your training dataset
    block_size=128  # Adjust the block size according to your dataset
)

model = GPT2LMHeadModel.from_pretrained('gpt2')
training_args = TrainingArguments(
    output_dir='/content/drive/MyDrive/nlp_final_model/', 
    overwrite_output_dir=True,
    num_train_epochs=8,  
    per_device_train_batch_size=32,  
    learning_rate=1e-5,  
    save_steps=300,    
    prediction_loss_only=True,
)
```

```python
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_dataset,
)

trainer.train()

# Step 4: Save the fine-tuned model
model.save_pretrained('/content/drive/MyDrive/Colab Notebooks/nlp_final_model/')
```
